```python
from stl import mesh
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
```

```python
# Load the STL files and add the vectors to the plot
your_mesh = mesh.Mesh.from_file("hbar-botched.stl")
good_mesh = mesh.Mesh.from_file("hbar-compressed.stl")
weird_mesh = mesh.Mesh.from_file("hbar-weird.stl")
```

```python
fig = plt.figure(figsize = (8, 3))
ax = fig.add_subplot(1, 3, 1, projection='3d')

poly_collection = mplot3d.art3d.Poly3DCollection(weird_mesh.vectors,
                                                 shade = False,
                                                 facecolors = "gray",
                                                 edgecolors = "k", lw = 0.01, antialiased=True)
ax.add_collection3d(poly_collection)

ax.set_box_aspect((1, 0.25, 1))
ax.axis('off')
ax.set_facecolor('none')

ax = fig.add_subplot(1, 3, 2, projection='3d')
ax.view_init(elev=20, azim=-90, roll=-20)
poly_collection = mplot3d.art3d.Poly3DCollection(good_mesh.vectors,
                                                 shade = False,
                                                 facecolors = "gray",
                                                 edgecolors = "k", lw = 0.01, antialiased=True)
ax.add_collection3d(poly_collection)
ax.set_box_aspect((1, 0.25, 1))
ax.axis('off')
ax.set_facecolor('none')

ax = fig.add_subplot(1, 3, 3, projection='3d')
ax.view_init(elev=180, azim=-50, roll=30)
poly_collection = mplot3d.art3d.Poly3DCollection(your_mesh.vectors,
                                                 shade = False,
                                                 facecolors = "gray",
                                                 edgecolors = "k", lw = 0.01, antialiased=True)
ax.add_collection3d(poly_collection)
ax.set_box_aspect((1, 0.25, 1))

ax.axis('off')
ax.set_facecolor('none')

fig.set_facecolor("#f0f0f0")
fig.tight_layout()
fig.savefig("hbars.png", dpi = 300)
```
