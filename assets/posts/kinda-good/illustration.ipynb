```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif"
})
```

```python
np.random.seed(87)

sig = np.random.rand(100) - 0.5
plt.plot(np.cumsum(sig))
```

```python
fig, ax = plt.subplots(figsize = (9,2.4))
fig.patch.set_facecolor('#f0f0f0')

sc = 20
o1 = 0.65
o2 = 0.1

xd = np.linspace(0.07, 0.97, 2000)
yd = 0.07*(np.sin(xd * 100) + 0.3*np.sin(xd * 200) + 0.5*np.sin(xd * 300)) * np.exp(-(xd-0.5)**2/0.4**2) + 0.65
yd2 = 0.07*(np.sin(xd * 100 + np.pi) + 0.3*np.sin(xd * 200) + 0.5*np.sin(xd * 300)) * np.exp(-(xd-0.5)**2/0.4**2) + 0.85

y3 = 0.07*(np.sin(xd * 50) + 0.3*np.sin(xd * 100) + 0.4*np.sin(xd*150)) * np.exp(-(xd-0.5)**2/0.4**2)  + 0.27
y4 = 0.07*(np.sin(xd * 50 + np.pi/2) + 0.3*np.sin(xd * 100) + 0.4*np.sin(xd * 150)) * np.exp(-(xd-0.5)**2/0.4**2) + 0.13

ax.fill_between(xd, yd, yd2, color = "gray")
ax.fill_between(xd, y3, y4, color = "gray")


ax.hlines(np.arange(5)/sc + o1, 0, 1, color = "k", lw = 0.5)
ax.hlines(np.arange(5)/sc + o2, 0, 1, color = "k", lw = 0.5)
# ax.vlines([0, 0.993, 1], o2, 4/sc+o2, color = "k")
ax.vlines([0, 0.993, 1], o2, 4/sc+o1, color = "k")

ax.text(0.01, 0.66, r"$\int$", fontsize = 45)
ax.text(0.01, 0.12, r"$\int$", fontsize = 45)

ax.set_xlim([0, 1])
ax.set_ylim([0, 1])

ax.set_xticks([])
ax.set_yticks([])
ax.patch.set_facecolor('#f0f0f0')
ax.axis('off')
fig.tight_layout()
fig.savefig("sheet.png", dpi = 300)
fig.savefig("sheet.svg")
```
